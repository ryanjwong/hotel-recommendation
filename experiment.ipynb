```python
blah['name_processed'] = blah['Hotels'].str.lower().str.split()
```

```python
# Convert 'name_processed' lists to strings for exact matching
blah['name_processed'] = blah['name_processed'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Find the intersection of unique hotel names
matching_hotels = set(blah['name_processed']).intersection(set(hotel_review_sample['Hotel_Name']))

# Count the number of matches
num_matches = len(matching_hotels)

print(f"Number of matching hotel names: {num_matches}")
```
